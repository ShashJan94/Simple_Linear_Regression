```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
import math
from pydoc import help
from scipy.stats.stats import pearsonr
```

```python
dataset = pd.read_csv("diamonds.csv")
```

```python
dataset
```

While performing simple linear regression- we consider only one independent variable. 
# An indpendent variable is such an entity which can regulate changes in the dependent variable. Conventionally, independant variables are denoted as 'X' and dependents as 'Y'

```python
#Declaration of the variables
X = dataset["depth"]
Y= dataset["price"]
```

```python
#Now to move onto simple linear regression-
#Add the bias to the X matrix
from statsmodels.api import add_constant
Xb=add_constant(X) #we are adding the intercept which is denoted as 'b' in the text books a.k.a - y= mx + b.
ln_fit = sm.OLS(Y,Xb)
fitted_model = ln_fit.fit()
fitted_model.summary()
```

```python
print((pearsonr(X,Y)[0])**2) #The R^2 or 1-(sum((Y-Y_hat)**2)/sum((Y-Y_mean)**2) in a formulated way. 
```

What is a R squared? Difference ratio of three variables Y, Y_hat, Y_mean. The larger the ratio of the distance substraction, the larger the distance ratio of the variables. If the variables are widely parted then correlation between independent variable and the dependent variable is low. We need to discard the explanatory variable.
